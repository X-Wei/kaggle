Define a useful function: plot scatters by 2 attributes:

```python
def plot_cov( c1, c2, cls=train.Cover_Type.values, figsize=(15,6) ):
    '''
    Plot cov-plots between columns, c1 in X-axis, c2 in Y-axis
    '''
    fig = plt.figure(figsize=figsize)
    plt.scatter(c1, c2, c=cls, s=100)
    plt.grid()
    plt.xlabel(c1.name)
    plt.ylabel(c2.name)
```

### 1. Aspect is in degrees azimuth(方位角), shift it at 180

**(目的是使得地理位置上接近的地点(如方位角=5和方位角=355的两处)在Aspect或Aspect2的某一个上接近)**

```python
def rotate180(x):
    return (x+180)%360

def rotate100(x):
    return (x+100)%360 # maybe rotate 100 deg is better: have less instances near 0 or 360...

train['Aspect2'] = train.Aspect.map(rotate180)
test['Aspect2'] = test.Aspect.map(rotate180)

train['Aspect3'] = train.Aspect.map(rotate100)
test['Aspect3'] = test.Aspect.map(rotate100)

train[['Aspect2','Aspect3']].hist(bins=50)
```

### 2. Create variable which indicates positive or negative value for `Vertical_Distance_To_Hydrology`

```python
train['Highwater'] = train.Vertical_Distance_To_Hydrology < 0
test['Highwater'] = test.Vertical_Distance_To_Hydrology < 0
```

### 4. Doing what Alexander Guschin suggests (*FE2*):

```python
train['Distanse_to_Hydrolody'] = (train['Horizontal_Distance_To_Hydrology']**2+\
                                            train['Vertical_Distance_To_Hydrology']**2)**0.5
test['Distanse_to_Hydrolody'] = (test['Horizontal_Distance_To_Hydrology']**2+\
                                            test['Vertical_Distance_To_Hydrology']**2)**0.5

train['Hydro_Fire_1'] = train['Horizontal_Distance_To_Hydrology']+\
                                            train['Horizontal_Distance_To_Fire_Points']
test['Hydro_Fire_1'] = test['Horizontal_Distance_To_Hydrology']+\
                                            test['Horizontal_Distance_To_Fire_Points']

train['Hydro_Fire_2'] = abs(train['Horizontal_Distance_To_Hydrology']-\
                                            train['Horizontal_Distance_To_Fire_Points'])
test['Hydro_Fire_2'] = abs(test['Horizontal_Distance_To_Hydrology']-\
                                            test['Horizontal_Distance_To_Fire_Points'])

train['Hydro_Road_1'] = abs(train['Horizontal_Distance_To_Hydrology']+\
                                            train['Horizontal_Distance_To_Roadways'])
test['Hydro_Road_1'] = abs(test['Horizontal_Distance_To_Hydrology']+\
                                            test['Horizontal_Distance_To_Roadways'])

train['Hydro_Road_2'] = abs(train['Horizontal_Distance_To_Hydrology']-\
                                            train['Horizontal_Distance_To_Roadways'])
test['Hydro_Road_2'] = abs(test['Horizontal_Distance_To_Hydrology']-\
                                            test['Horizontal_Distance_To_Roadways'])

train['Fire_Road_1'] = abs(train['Horizontal_Distance_To_Fire_Points']+\
                                            train['Horizontal_Distance_To_Roadways'])
test['Fire_Road_1'] = abs(test['Horizontal_Distance_To_Fire_Points']+\
                                            test['Horizontal_Distance_To_Roadways'])

train['Fire_Road_2'] = abs(train['Horizontal_Distance_To_Fire_Points']-\
                                            train['Horizontal_Distance_To_Roadways'])
test['Fire_Road_2'] = abs(test['Horizontal_Distance_To_Fire_Points']-\
                                            test['Horizontal_Distance_To_Roadways'])
```

### 1. Adding more info about soil types

```python
SoileType_addCols = [
        ["Rock outcrop complex" , 1, 3, 4, 5, 6, 10, 27, 28, 33], 
        ["Ratake families",2,4], 
        ["extremely stony", 1, 24, 25, 27,28,29,30,31, 32, 33, 34, 36, 37, 38, 39, 40], 
        ["very stony", 2, 9, 18, 26], 
        ["rubbly", 3, 4, 5, 10, 11, 13], 
        ["Vanet",2,5,6],
        ["stony",6,12],
        ["Bullwark",10, 11, 13 ],
        ["Catamount families",10,11,13,26,31,32,33],
        ["Rock land",11,12,13,30,34,36,40],
        ["Cryaquolis",16,17],
        ["typic cryaquolls",20,21,23],
        ["Leighcan",21,22,23,24,25,27,28,31,32,33,38,39],
        ["till substratum",21,22,23],
        ["warm",25,27],
        ["Como",29,30],
        ["Legault",29,30],
        ["Cryorthents",34,37,39,40],
        ["Cryumbrepts",35,36,37],
        ["Moran families",38,39,40],
    ]
```

**Adding these informations:**

```python
def f( row, col_names ):
    return row[col_names].sum() # returns 1 or 0

for ci,cols in list( enumerate(SoileType_addCols) ):
    print '%d out of %d......'% ( ci+1, len(SoileType_addCols) )
    
    new_col_name = cols[0] # name for this new columns
    col_names = ['Soil_Type%d'%i for i in cols[1:] ]

    new_col = train.apply( lambda r:f(r, col_names), axis=1 ) #since f takes 2 arguments, use lambda
    # we can also write: 
    # new_col = train.apply( f, axis=1, args=(col_names,) )
    train[new_col_name] = new_col
    
    new_col = test.apply(lambda r:f(r, col_names), axis=1)
    test[new_col_name] = new_col
    
```
